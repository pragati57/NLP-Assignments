```python
#7

def countDog(st):
  count=0
  for word in st.lower().split():
    if word=='dog':
      count+=1
  return count

countDog('This dog runs faster than other dog !')
```

```python
#9

def caught_speeding(speed, is_birthday):
    
    if is_birthday:
        speeding = speed - 5
    else:
        speeding = speed
    
    if speeding > 80:
        return 'Heavy Challan'
    elif speeding > 60:
        return 'Small Challan'
    else:
        return 'No Challan'

caught_speeding(85,True)
```
